```python
model = genai.GenerativeModel("gemini-2.5-flash")
```

```python
historico = """
  Você é um especialista em educação financeira e finanças pessoais.
  Seu objetivo é ajudar pessoas com dúvidas ou problemas financeiros.

  Diretrizes importantes:
  - Faça uma análise da situação ou questão financeira da pessoa, sem emitir julgamentos sobre a pessoa.
  - Procure dar conselhos e orientações personalizadas nas suas respostas.
  - Utilize palavras simples e comuns. Seja didático nas explicações e orientações dadas.
  """
```

```python
nome = input("\nDigite seu nome completo: ")

while True:

  usuario = input(f"\n{nome}, relate sua dúvida ou situação/problema financeiro (ou digite sair): ")

  if usuario.lower() == "sair":
    print("Foi bom conversar contigo! Caso tenha mais alguma dúvida ou questão financeira, me mande uma nova mensagem!!")
    break

  historico += f"\nUsuário {nome}: {usuario}"

  response = model.generate_content(historico)
  resposta = response.text

  print("\nEducador Financeiro:")
  print(resposta)

  historico += f"\nEducador Financeiro: {resposta}"
```
